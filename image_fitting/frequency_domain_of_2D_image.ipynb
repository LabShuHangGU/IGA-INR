```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
from numpy.fft import fft2, fftshift
```

```python
# Read the ground truth image and its corresponding implicit neural representation.
# Transfer to the gray image.
gt_img = io.imread('results/kodim05.png', as_gray=True)
gt_img = gt_img.astype(np.float32)

# (x, y) denotes the position of top left corner of the patch.
top_left = (152, 205)  
patch_size = 50  

# Crop the corresponding patch.
gt_patch = gt_img[top_left[1]:top_left[1]+patch_size, top_left[0]:top_left[0]+patch_size]
```

```python
inr_1 = io.imread('results/relu+pe_kodim05_24.60.png', as_gray=True)
inr_1 = inr_1.astype(np.float32)
inr_patch_1 = inr_1[top_left[1]:top_left[1]+patch_size, top_left[0]:top_left[0]+patch_size]
# Compute the residual patch.
residual_patch_1 = gt_patch - inr_patch_1
# Perform a 2D Fourier transform on the residual patch.
F_residual = fft2(residual_patch_1)
# Shift the zero-frequency component to the center.
F_residual_shifted = fftshift(F_residual)  
# Take the logarithm to compress the dynamic range.
magnitude_spectrum_residual = np.log(np.abs(F_residual_shifted) + 1)

# Visualization
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(residual_patch_1, cmap='gray')
plt.title('Residual Image')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(magnitude_spectrum_residual, cmap='jet')  
plt.title('Frequency Domain of Residual by PE')
plt.colorbar(label='Magnitude', shrink=0.7)
plt.axis('off')
plt.gcf().patch.set_alpha(0) 
plt.show()
```
